````
# a dictionary-based word count function
def word_counts_dict(long_txt):
  words = long_txt.lower().replace("\n","").replace(".","").replace(",","").replace("—"," ").split()
  wc = {}
  for word in words:
    if word not in wc:
      wc[word] = 1
    else:
      wc[word] +=  1
  return wc

word_counts_dict(gettysburg)
````

````
def word_counts_list(long_txt):
  '''
  function returns the total count of each unique word in the input string
  '''

  words = long_txt.lower().replace("\n","").replace(".","").replace(",","").replace("—"," ").split() # takes our input doc string and converts it to a list of every word 
  print(words) # check the above methods transformed the doc string to a list of words
  unique_words = [] # accumulator initialized for each unique word in input string
  counts = [] # accumulator initialized for updating the total count of each unique word appears in input string

  # iterate through each word in our list of words
  for word in words: 
      if word not in unique_words: 
          unique_words += [word] # if our unique word is not yet in accumulator for unique_words, add the word to the list
          counts += [1] # an update our accumulator for total count of each unique word by 1
      else: 
          counts[unique_words.index(word)] += 1 # if unique word already appears in unique_words, in the current position of the unique word, add 1 to count
  return list(zip(unique_words, counts)) # returns a single list containing a tuple of the unique_word and its counts

word_counts_list(gettysburg)
````

````
#@title Click here for a solution

def roman2int(roman_numeral):
    rules = {"IV":4, "IX":9,"XL":40,"XC":90,"CD":400,"CM":900,
             "I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000}
    
    # initialize loop variables
    remaining = str(roman_numeral)
    int_value = 0
    
    while remaining: 
        # check all two-character rules
        if remaining[:2] in rules:
            int_value += rules[remaining[:2]]
            remaining = remaining[2:]
            
        # check all one-character rules    
        elif remaining[0] in rules:
            int_value += rules[remaining[0]]
            remaining = remaining[1:]
        # if no rule matches then consume the first character
        else:
            remaining = remaining[1:]
            
    return int_value

print(roman2int("XXIX"))
print(roman2int("MCMXCIX"))
print(roman2int("Super Bowl XXXIV"))
````
